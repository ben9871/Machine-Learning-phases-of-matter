```python
import numpy as np
import matplotlib.pyplot as plt


L=[]
with open("lattice_sizes.txt") as f:
        L = f.readlines()
modified_strings = [s.replace('e+01\n', '') for s in L]
integers = [int(float(s)) for s in modified_strings]
result = ', '.join(str(i) for i in integers)
#print(result)
L=[]
for i in (integers):
    L.append(i*10)

# import lattices text file

Xtests = []
Xtrains = []
ytests = []
ytrains = []
Ytrains = []
Ytests = []

#empty lists to which we shall append the training and testing data
for k, i in enumerate(L):

    a = np.load('X_tests_{} .npz'.format(i))
    Xtests.append(a['lats'])
    
    b = np.load('X_trains_{} .npz'.format(i))
    Xtrains.append(b['lats'])
    
    c = np.load('y_tests_{} .npz'.format(i))
    ytests.append(c['onehots'])
    
    d =np.load('y_trains_{} .npz'.format(i))
    ytrains.append(d['onehots'])
    
    e =np.load('Yor_trains_{} .npz' .format(i))
    Ytrains.append(e['temps'])
    
    f =np.load('Yor_tests_{} .npz' .format(i))
    Ytests.append(f['temps'])

```

```python
def accuracy(yp, y):
    N = y.shape[0] # number of examples
    acc = (1./N) * np.trace(np.dot(yp,y.T))
    return acc

def heaviside(X):
    X[X >= 0] = 1
    X[X < 0] = 0
    return X
        
def toy_model(X, eps): # apply to a NxLxL stack of images
    m = X.shape[1]*X.shape[2]
    X = X.reshape((-1,m)) 
    
    W1 = (1./m) * np.stack((np.ones((m,)),-np.ones((m,)), np.ones((m,))))
    W1 = (1./(1.+eps)) * W1
    b1 = np.array([-1,-1, 1])
    b1 = (eps/(1.+eps)) * b1
    
    h1 = np.dot(X, W1.T) + b1
    h1 = heaviside(h1)
    
    W2 = np.stack(([2, 1, -1], [-2, -2, 1]))
    
    h2 = np.dot(h1, W2.T)
    y = heaviside(h2)
    
    return y 
```

```python
epsilons = np.arange(0.00, 1.00, 0.01)
accuracies = np.zeros(epsilons.shape[0])
eps_best_L = []
# try out all possible epsilons and see how they fare
#make empty list which will hold the best epsilon for each lattice size


for k, j in enumerate(L):
    for i, eps in enumerate(epsilons):
        print(i,'/', len(epsilons))
        print('epsilon:', eps)
        accuracies[i] = accuracy(toy_model(Xtrains[k], eps), ytrains[k])
        print('train set accuracy{}:'.format(j), accuracies[i])
    # probe accuracies
    i_best = np.argmax(accuracies)
    eps_best = epsilons[i_best]
    train_accuracy_best = accuracies[i_best]
    eps_best_L.append(eps_best)
    #append to list
    print('Best train set epsilon{}: '.format(j) + str(eps_best))
    print('Best train set accuracy{}: '.format(j) + str(train_accuracy_best))



fig, ax = plt.subplots(figsize=(15,10))

#plot accuracies as a function of epsilon

ax.plot(epsilons, accuracies, marker = 'o', color = 'r')
ax.set_xlabel('Epsilon')
ax.set_ylabel('Accuracy')
ax.set_title('Train Set Accuracy vs. Epsilon for Toy Model')
ax.grid(color='b', linewidth=.2)
plt.show()
fig.savefig('output/accuracy_v_epsilon_toy_model.jpg')
```
